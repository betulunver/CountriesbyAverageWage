### Import Libraries

```python
import pandas as pd
import bar_chart_race as bcr
```

```python
df=pd.read_csv("avg_wage.csv")
```

```python
df.sample(5)
```

```python
df.isnull().sum() # Boş veri yok.
```

```python
df2=df.transpose() # satır ve sütunlar yer değiştirdi.
```

```python
df2
```

```python
df2.rename(columns=df2.iloc[0], inplace = True) # header'ı sildik.
```

```python
df2 = df2.drop(df2.loc[df2.index=='Country'].index)
```

```python
df2
```

```python
df2.index.name = 'Year' # index ismini 'Year' yaptık.
```

```python
df2
```

```python
df_total=df2.cumsum(axis=0) # DataFrame ekseni üzerinden kümülatif toplamı döndür.
```

```python
df_total = df_total.astype('int') #kolonlar int' çevirdik.
```

```python
bcr.bar_chart_race(df_total,filename="countriesbyavaragewage.mp4",figsize=(8,5),title="Countries By Average Wage")
```
